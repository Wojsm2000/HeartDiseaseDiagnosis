```python
with open("data/new.data", "r") as file:
    lines = file.readlines()


cleaned_data = []
current_row = []
for line in lines:

    values = line.strip().split()
    print(values)
    current_row.extend(values)

    if len(current_row) >= 90:
        cleaned_data.append(current_row[:90])
        current_row = current_row[90:]


with open("cleaned_file.data", "w") as file:
    for row in cleaned_data:
        file.write(" ".join(row) + "\n")
```
